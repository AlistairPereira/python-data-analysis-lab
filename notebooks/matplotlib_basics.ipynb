```python
x=np.arange(1,10)
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(2,2,1)
ax1.plot(x,x*x,'ro-')
ax1.set_title('Graph A')

ax2 = fig.add_subplot(2,2,3)
ax2.plot(x,x*3,'yv--')
ax2.set_title('Graph B')

ax3 = fig.add_subplot(2,2,2)
ax3.plot(x,x*1.5,'go--')
ax3.set_title('Graph C')

ax4 =fig.add_subplot(2,2,4)
ax4.hist(x, color="blue", edgecolor="yellow")
ax4.set_title("Graph D")
```

```python
programming_language=['python','java','c','Go','Ruby']
no_of_students=[50,33,27,35,25]
```

```python
figure = plt.figure(figsize=(10,5))
plt.bar(programming_language,no_of_students, color="yellow")
for i,v in enumerate(no_of_students):
    plt.text(i,v,str(v),ha='center',va='bottom')
plt.xlabel("lang")
plt.ylabel('students')
plt.title("language stats")
figure.savefig('bar.jpg')
plt.show()
```

```python
#Pie and doghnut chart
```

```python
quantity =np.array([30,30,30,10])
fruits =['apple','orange','lemon','pineapple']
```

```python
plt.pie(quantity,labels = fruits,autopct='%1.2f%%')
plt.show()
```

```python
my_explode=[0,0,0.3,0]
plt.pie(quantity,labels=fruits, autopct='%1.2f%%',explode = my_explode)
plt.show()
```

```python
grades=[1,2,3,4,5,6,7,8]
b_grades=[10,20,30,40,50,60,70,80]
students=[10,12,45,23,56,67,7,77]
```

```python
plt.scatter(students,grades, color='b',marker='o')
plt.scatter(students, b_grades, color='r',marker ="*")
plt.xlabel("students")
plt.ylabel("grades")
plt.show()
```

```python
plt.plot(students,grades, color='y',marker='o')
plt.plot(students, b_grades, color='r',marker ="*")
plt.xlabel("students")
plt.ylabel("grades")
plt.show()
```
