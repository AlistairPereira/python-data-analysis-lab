```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
    {'name':['mohan','sara','sakib','tao','virat','alistair','amroy','rahul','aaron'],
     'height':[1.2,2.3,4.9,5.1,5.2,3.2,2.9,14.5,5.8]}
)
```

```python
df
```

```python
q1= df['height'].quantile(0.25)
q1
```

```python
q3 = df['height'].quantile(0.75)
q3
```

```python
iqr = q3-q1
iqr
```

```python
lower_limit = q1-1.5*iqr
lower_limit
```

```python
upper_limit = q3 + 1.5 * iqr
upper_limit
```

```python
outlier_data = df[(df['height'] < lower_limit) | (df['height'] >= upper_limit)]
outlier_data
```

```python
non_outlier_data = df[(df['height']> lower_limit) & (df['height']< upper_limit)]
non_outlier_data
```

```python
#2nd method standard deviation
```

```python
df['height'].mean()
```

```python
df['height'].std()
```

```python
upper_limit_std = df['height'].mean()+3*df['height'].std()
upper_limit_std
```

```python
lower_limit_std = df['height'].mean() -3 * df['height'].std()
lower_limit_std
```

```python
outlier_data_std = df[(df['height'] < lower_limit_std) | (df['height'] >= upper_limit_std)]
outlier_data_std
```

```python
#method Z score z= x-u(mean)/std_dev
```

```python
df['height'].mean()
```

```python
df['height'].std()
```

```python
df['zscore_height']= (df['height']-df['height'].mean())/df['height'].std()
df['zscore_height']
```

```python
df['zscore_height'].max()  #upper limit
```

```python
df['zscore_height'].min() #lower limit
```

```python
outlier_data_zscore = df['height'] > df['zscore_height'].max()
outlier_data_zscore
```
