```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.lineplot(x='size',y='total_bill',data=tips)
plt.show()
```

```python
plt.plot(tips['size'],tips['total_bill'],color="red")
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
tips['tip'].mean()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex')
plt.show()
#categorize on the bases of sex (hue)
#food id eaten more by males than femals  & tips are given more by males than femals
# the more the bill the more tip given
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data =tips)
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='time',data =tips)
plt.show()
#more tip and bill are at dinner time 
```

```python
my_markers = {'male':'-','female':'x'}
sns.scatterplot(x='total_bill',y='tip',hue='sex', data=tips,markers=my_markers,legend=False)
plt.show()
```

```python
#relationplot
sns.relplot(x='total_bill',y='tip',data=tips, hue ='smoker',col='time')
plt.show()
```
